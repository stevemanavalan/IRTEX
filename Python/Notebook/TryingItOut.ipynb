```python
import pandas as pd
```

```python
# Read dataset feature vectors
data_features = pd.read_csv('outputFeature.csv', header=None)
data_features
```

```python
# Read query image feature vector
query_feature = pd.read_csv('queryFeature.csv', header=None)
query_feature
```

```python
# Dropping 1st column from the dataframe
data_feature_vector = data_features.drop(columns=[0])
data_feature_vector
```

## COSINE SIMILARITY BETWEEN QUERY AND DATASET 

```python
# cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
cos_sim = cosine_similarity(query_feature_vector, data_feature_vector)
cos_sim
```

```python
cos_sim = cos_sim.flatten()
cos_sim
```

```python
# Create cosine similarity values dataframe
cos_sim_df = pd.DataFrame(cos_sim)
cos_sim_df
```

```python
cos_sim_df_new = pd.DataFrame(cos_sim, columns=['cos_sim'])
cos_sim_df_new
```

```python
# Add cosine similarity as lasy column in data feature dataframe as ground truth
data_features_y = pd.concat([data_features, cos_sim_df], axis=1)
data_features_y
```

## CREATING A NEW CSV FILE WITH COSINE SIMILARITY AS LAST COLUMN

```python
# Creating output csv file with cosine similarity
data_features_y.to_csv('out1.csv', index=False, header=False)
```
